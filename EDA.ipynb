```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned_agriculture_data.csv")
```

```python
df.head()
```

```python
## Top 7 RICE PRODUCTION State Data(Bar_plot)

top_rice_states = df.groupby("state_name")["rice_production_1000_tons"].sum().sort_values(ascending=False).head(7)
```

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)

plt.figure(figsize=(10,6))
top_rice_states.plot(kind="bar", color="green")
plt.title("Top 7 Rice Producing States")
plt.xlabel("State")
plt.ylabel("Total Rice Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

top_wheat_states = df.groupby("state_name")["wheat_production_1000_tons"].sum().sort_values(ascending=False).head(5)
```

```python
plt.figure(figsize=(8,8))
top_wheat_states.plot(
    kind="pie",
    autopct='%1.1f%%',  # Shows percentage with 1 decimal place
    startangle=90,      # Rotate so the first slice starts at the top
    legend=True
)
plt.title("Top 5 Wheat Producing States (Percentage Share)")
plt.ylabel("")  # Remove y-axis label for cleaner pie chart
plt.tight_layout()
plt.show()
```

```python
#Oil seed production by top 5 states
top_oilseed_states = df.groupby("state_name")["oilseeds_production_1000_tons"].sum().sort_values(ascending=False).head(5)


```

```python
#Oil seed production by top 5 states

# Bar chart for Oilseed production
plt.figure(figsize=(10,6))
top_oilseed_states.plot(kind="bar", color="olive")
plt.title("Top 5 Oilseed Producing States")
plt.xlabel("State")
plt.ylabel("Total Oilseed Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie chart for percentage share
plt.figure(figsize=(8,8))
top_oilseed_states.plot(
    kind="pie",
    autopct='%1.1f%%',
    startangle=90,
    legend=True
)
plt.title("Oilseed Production Share - Top 5 States")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
recent_years = df['year'].max() - 50
df_last_50_years = df[df['year'] >= recent_years]
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
sugarcane_trend = df_last_50_years.groupby('year')['sugarcane_production_1000_tons'].sum()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

plt.figure(figsize=(12,6))
plt.plot(sugarcane_trend.index, sugarcane_trend.values, marker='o', linestyle='-', color='brown')
plt.title("India's Sugarcane Production (Last 50 Years)")
plt.xlabel("Year")
plt.ylabel("Total Sugarcane Production (1000 tons)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Rice Production Vs Wheat Production (Last 50y)

recent_years = df['year'].max() - 50
df_last_50_years = df[df['year'] >= recent_years]
```

```python
rice_production = df_last_50_years.groupby('year')['rice_production_1000_tons'].sum()
wheat_production = df_last_50_years.groupby('year')['wheat_production_1000_tons'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(rice_production.index, rice_production.values, label='Rice Production', marker='o', color='green')
plt.plot(wheat_production.index, wheat_production.values, label='Wheat Production', marker='s', color='orange')

plt.title("Rice vs Wheat Production in India (Last 50 Years)")
plt.xlabel("Year")
plt.ylabel("Total Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Millet Production (Last 50y)
## Filter data for the last 50 years
recent_years = df['year'].max() - 50
df_last_50_years = df[df['year'] >= recent_years]
```

```python
# Group by year and sum both millet columns
millet_trend = df_last_50_years.groupby('year')[[
    'pearl_millet_production_1000_tons', 
    'finger_millet_production_1000_tons'
]].sum()

```

```python
# Add the two columns together to get total millet production
millet_trend['total_millet_production_1000_tons'] = (
    millet_trend['pearl_millet_production_1000_tons'] + 
    millet_trend['finger_millet_production_1000_tons']
)
```

```python
# Plot total millet production
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(millet_trend.index, millet_trend['total_millet_production_1000_tons'], marker='o', color='purple')
plt.title("Millet Production in India (Last 50 Years)")
plt.xlabel("Year")
plt.ylabel("Total Millet Production (1000 tons)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by state (region) and sum kharif and rabi sorghum production
sorghum_by_region = df.groupby('state_name')[
    ['kharif_sorghum_production_1000_tons', 'rabi_sorghum_production_1000_tons']
].sum()

```

```python
# Plot stacked bar chart
sorghum_by_region_sorted = sorghum_by_region.sort_values(by='kharif_sorghum_production_1000_tons', ascending=False)

sorghum_by_region_sorted.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6),
    color=['#FFA07A', '#20B2AA']
)

plt.title("Sorghum Production (Kharif and Rabi) by Region")
plt.xlabel("State (Region)")
plt.ylabel("Sorghum Production (1000 tons)")
plt.xticks(rotation=90)
plt.legend(["Kharif Sorghum", "Rabi Sorghum"])
plt.tight_layout()
plt.show()
```

```python
#Soybean Production by Top 5 States and Yield Efficiency
# Group by state and calculate total soybean production and average yield
soybean_stats = df.groupby('state_name').agg({
    'soyabean_production_1000_tons': 'sum',
    'soyabean_yield_kg_per_ha': 'mean'
})
```

```python
# Get top 5 states by soybean production
top5_soybean = soybean_stats.sort_values(by='soyabean_production_1000_tons', ascending=False).head(5)
```

```python
# Plot bar chart for production
plt.figure(figsize=(8,4))
top5_soybean['soyabean_production_1000_tons'].plot(kind='bar', color='goldenrod')
plt.title("Top 5 Soybean Producing States")
plt.xlabel("State")
plt.ylabel("Soybean Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot bar chart for yield efficiency
plt.figure(figsize=(7,3))
top5_soybean['soyabean_yield_kg_per_ha'].plot(kind='bar', color='seagreen')
plt.title("Yield Efficiency of Top 5 Soybean Producing States")
plt.xlabel("State")
plt.ylabel("Average Yield (Kg per ha)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
import seaborn as sns
```

```python
# Set up the figure
plt.figure(figsize=(15, 5))

# Scatter Plot: Rice
plt.subplot(1, 3, 1)
sns.scatterplot(
    x='rice_area_1000_ha',
    y='rice_production_1000_tons',
    data=df,
    color='green'
)
plt.title('Rice: Area Cultivated vs Production')
plt.xlabel('Area Cultivated (1000 ha)')
plt.ylabel('Production (1000 tons)')

# Scatter Plot: Wheat
plt.subplot(1, 3, 2)
sns.scatterplot(
    x='wheat_area_1000_ha',
    y='wheat_production_1000_tons',
    data=df,
    color='orange'
)
plt.title('Wheat: Area Cultivated vs Production')
plt.xlabel('Area Cultivated (1000 ha)')
plt.ylabel('Production (1000 tons)')

# Scatter Plot: Maize
plt.subplot(1, 3, 3)
sns.scatterplot(
    x='maize_area_1000_ha',
    y='maize_production_1000_tons',
    data=df,
    color='blue'
)
plt.title('Maize: Area Cultivated vs Production')
plt.xlabel('Area Cultivated (1000 ha)')
plt.ylabel('Production (1000 tons)')

plt.tight_layout()
plt.show()
```

```python
#Rice vs. Wheat Yield Across States

# Group by state and calculate average yield for rice and wheat
yield_by_state = df.groupby('state_name')[
    ['rice_yield_kg_per_ha', 'wheat_yield_kg_per_ha']
].mean()
```

```python
# Plot bar chart for Rice vs Wheat Yield
yield_by_state.sort_values('rice_yield_kg_per_ha', ascending=False, inplace=True)
yield_by_state.plot(
    kind='bar',
    figsize=(14,7),
    color=['green', 'orange']
)

plt.title('Rice vs Wheat Yield Across States')
plt.xlabel('State')
plt.ylabel('Average Yield (Kg per ha)')
plt.xticks(rotation=90)
plt.legend(['Rice Yield', 'Wheat Yield'])
plt.tight_layout()
plt.show()
```
